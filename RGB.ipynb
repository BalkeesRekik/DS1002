```python
import numpy as np
from PIL import Image
import requests
from io import BytesIO
import matplotlib.pyplot as plt #used this for plotting (see citation)


def red_filter(photo):
  red_only = np.copy(photo)
  red_only[: , : , 1:] = 0
  return red_only

def green_filter(photo):
  green_only = np.copy(photo)
  green_only[: , : , ::2] = 0
  return green_only

def blue_filter(photo):
  blue_only = np.copy(photo)
  blue_only[:, :, :2] = 0
  return blue_only

def red_green_filter(photo):
  red_green_only = np.copy(photo)
  red_green_only[:, :, 2] = 0
  return red_green_only

def red_blue_filter(photo):
  red_blue_only = np.copy(photo)
  red_blue_only[:, :, 1] = 0
  return red_blue_only

def green_blue_filter(photo):
  green_blue_only = np.copy(photo)
  green_blue_only[:, :, 0] = 0
  return green_blue_only

photo_url = input("Please enter a photo's URL: ")
response = requests.get(photo_url)
photo = np.array(Image.open(BytesIO(response.content))) #converted the photo into a np array

filtered_photos = [red_filter(photo), green_filter(photo), blue_filter(photo),
                  red_green_filter(photo), red_blue_filter(photo), green_blue_filter(photo)]

num_filters = 7

my_figure, subplots = plt.subplots(1, num_filters, figsize=(15,5)) #created a figure with subplots to diplays the pictures

subplots[0].imshow(photo)
subplots[0].set_title("Original")

for i in range(1, num_filters):
  title = ""
  if i==1:
    title = "Red Filter"
  elif i==2:
    title = "Green Filter"
  elif i==3:
    title = "Blue Filter"
  elif i==4:
    title = "Red & Green"
  elif i==5:
    title = "Red & Blue"
  elif i==6:
    title = "Green & Blue"

  subplots[i].imshow(filtered_photos[i-1])
  subplots[i].set_title(title)

plt.tight_layout() #citation
plt.show()

combined_filtered_photos = red_filter(photo) + green_filter(photo) + blue_filter(photo)

if np.array_equal(photo, combined_filtered_photos):
  print("Original photo is identical to combined filtered photos. ")
else:
  print("Original photo is not identical to combined filtered photos. ")





```
