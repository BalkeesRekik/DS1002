```python
!pip install english_words
from english_words import get_english_words_set
words = get_english_words_set(['web2'], lower=True)
import numpy as np
import matplotlib.pyplot as plt

wordlist = list(words)
print("The length of wordlist is: " , len(wordlist))

first_ten_words = wordlist[:10]
print(first_ten_words) # I noticed that the words are not alphabetically ordered.

alpha_list = sorted(wordlist)
print(alpha_list)

words_starting_with_a = 0
for word in alpha_list:
  if word[0] == "a":
    words_starting_with_a += 1
print("The number of words starting with 'a' : ", words_starting_with_a)

words_containing_q = 0
for word in alpha_list:
  if "q" in word:
    words_containing_q += 1
print("The number of words containing 'q' : ", words_containing_q)

words_not_containing_e = 0
for word in alpha_list:
  if "e" not in word:
    words_not_containing_e += 1
print("The number of words not containg 'e' : ", words_not_containing_e )

fraction = words_not_containing_e / 10
print("The fraction of words not containing 'e' : ", fraction)

alpha_array = np.array(alpha_list)
alphabet = [chr(i) for i in range(ord('a'),ord('z')+1)]
print(alphabet) # it generates the alphabets lowercase.

alphabet_num = {}
for word in alpha_list:
  first_letter = word[0]
  if first_letter in alphabet_num:
    alphabet_num[first_letter] += 1
  else:
     alphabet_num[first_letter] = 1

print("Number of words starting with each letter: ", alphabet_num) # printed the list so that it's easier for me to visualize it

letters = list(alphabet_num.keys())
numbers = list(alphabet_num.values())

plt.bar(letters, numbers)

plt.xlabel("Letters")
plt.ylabel("Numbers")
plt.title("Numbers of words starting with each letter")

plt.show()

longest_words = sorted(alpha_list, key=len, reverse=True)[:3]
print("The three longest words are: ", longest_words)

word_length = [len(word) for word in alpha_list]
max_length = max(word_length)
min_length = min (word_length)

length_frequency = {}
for length in range(min_length , max_length + 1):
  frequency = sum(1 for word in alpha_list if len(word) == length)
  length_frequency[length] = frequency

plt.bar(length_frequency.keys(), length_frequency.values())

plt.xlabel("Word length in letters")
plt.ylabel("Frequency")
plt.title("Histogram of word lengths")

plt.show()

most_common_length = max(length_frequency, key=length_frequency.get)
print("the most common length is: ", most_common_length)

words_containing_v_and_a = 0
for word in alpha_list:
  if "v" in word and "a" in word:
    words_containing_v_and_a += 1
print("The number of words containg both letters 'v' and 'a' : ",words_containing_v_and_a )

words_containing_w_and_l = 0
for word in alpha_list:
  if "w" in word and "l" in word:
    words_containing_w_and_l += 1
print("The number of words containg both letters 'w' and 'l' : ",words_containing_w_and_l )

# You have to loop through the word list once for each condition. The code will check every word in the full list to check
# if the condition is met (e.g. v and a).

# The whole code runs in 9s.


# BONUS

pairs_list = [] # Created a list that contains all the pairs possible

for letter1 in alphabet:
  for letter2 in alphabet:
    pairs_list.append(letter1 + letter2)

pair_freq = {}

for pair in pairs_list:  # Set the frequency for all pairs to 0
  pair_freq[pair] = 0

for word in alpha_list:
  for pair in pairs_list:
    if pair in word:
        pair_freq[pair] += 1

max_pair = max(pair_freq, key=pair_freq.get)
print("The pair of letters that appear the most is:", max_pair, ". It appears in " ,
pair_freq[max_pair], "words. ")

min_pair = min(pair_freq, key=pair_freq.get)
print("The pair of letters that appear the least is:", min_pair, ". It appears in " ,
pair_freq[min_pair], "words. ")

# The whole code runs in 29s now

heat_map = np.zeros((26,26))
# I gave up on the second part of the bonus question. It's quite hard!






```
